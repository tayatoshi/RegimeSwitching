```python
#np.random.seed(123)
"""
length = 600
trance = np.array([[0.85,0.15],
                   [0.2,0.8]])
tr = np.zeros(length)
y = np.zeros(length)
for i in range(1,length):
    if tr[i-1] == 0:
        tr[i]=np.random.binomial(1,trance[0,1])
    if tr[i-1] == 1:
        tr[i]=np.random.binomial(1,trance[1,1])
c = tr*30 + 30
y[0]=30
for i in range(1,length):
    y[i] = c[i]  + 0.7 * (y[i-1] - c[i-1])+ np.random.normal()
y=y.reshape(length,1)+np.random.normal(0,1,length).reshape(length,1)
a0 = np.array([30,60]).reshape(2,1)
p0 = np.ones([2,1,1])
"""
length = 200
a =np.zeros(length)
a[0]=30
for i in range(length):
    if i < 50:
        a[i] = 30  + 0.6 * (a[i-1] - 30) + np.random.normal()
    elif i < 100:
        a[i] = 60 + 0.6 * (a[i-1] - 60)+ np.random.normal()
    elif i < 150:
        a[i] = 30  + 0.6 * (a[i-1] - 30) + np.random.normal()
    else:
        a[i] = 60 + 0.6 * (a[i-1] - 60)+ np.random.normal()
y = (a + np.random.normal(0,1,length)*2).reshape(length,1)
a0 = np.array([30,60]).reshape(2,1)
p0 = np.ones([2,1,1])

plt.plot(y)
plt.show()
```

```python
model = resim.Regimeswitching(y=y,a0=a0,p0=p0)
result = model.fit()
print(result)
```

```python
for i in range(len(result['marginal_probability'])):
    result['marginal_probability'][i] = result['marginal_probability'][i]/np.sum(result['marginal_probability'][i])
```

```python
plt.plot(result['marginal_probability'][:,0])
plt.show()
```

```python
length = len(y)
plt.figure(figsize=(12,9))
plt.subplot(3,1,1)
plt.plot(y,label="y")
plt.plot(result["marginal_a"][:,0],label="a0")
plt.plot(result["marginal_a"][:,1],label="a1")
plt.title("observation and state variables")
plt.legend()

plt.subplot(3,1,2)
probabi = np.zeros(length)
for i in range(length):
    if result['marginal_probability'][i,0]<result['marginal_probability'][i,1]:
        probabi[i] = result["marginal_a"][i,1]
    else:
        probabi[i] = result["marginal_a"][i,0]
plt.plot(y,alpha=0.5,label='y')
plt.plot(probabi,label='a')
plt.legend()

plt.subplot(3,1,3)
plt.plot(result["marginal_probability"][:,1])
plt.title("Pr(a_1|omega_t)")
plt.fill_between(range(length),0,result["marginal_probability"][:,1],facecolor='gray',alpha=0.5)
plt.legend()
plt.show()
```
